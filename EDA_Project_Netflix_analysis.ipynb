# **Netflix Data Set Analysis**

* Netflix is one of the most popular media and video streaming platforms. They have over 8000 movies or tv shows available on their platform, as of mid-2021, they have over 200M Subscribers globally.
* This tabular dataset consists of listings of all the movies and tv shows available on Netflix, along with details such as - cast, directors, ratings, release year, duration, etc.





### **Description of each column in the dataset:**

**show_id:** A unique identifier for each show or movie.

**type:** The type of content, either "Movie" or "TV Show."(**object type**)

**title:** The title of the movie or TV show.(**object type**)

**director:** The director of the movie or TV show. In the first and third entries, this information is not available (NaN).(**object type**)

**cast:** The cast or actors in the movie or TV show. In the first entry, this information is not available (NaN). In the second entry, there is a list of actors from the TV show "Blood & Water."(**object type**)

**country:** The country where the movie or TV show was produced or is associated with.(**object type**)

**date_added**: The date when the content was added to the streaming platform, in the format "Month Day, Year."(**object type**)

**release_year:** The year the movie or TV show was originally released.(**int type**)

**rating:** The content's rating, which indicates the recommended audience age or maturity level (e.g., "PG-13" or "TV-MA").(**object type**)

**duration:** The duration of the movie or TV show. In the first entry, the duration is given in minutes ("90 min"). In the second and third entries, it's indicated in the number of seasons ("2 Seasons" and "1 Season").(**object type**)

**listed_in:** The genre or category of the content, which can help classify it (e.g., "Documentaries," "International TV Shows," "Crime TV Shows").(**object type**)

**description:** A brief description or synopsis of the movie or TV show, providing an overview of the plot or subject matter.(**object type**)

**7) Load the Dataset and find the measures of central tendency and dispersion of duration column by extracting  numerical part of the duration column.**

```python
df_Net_New = df_Net.copy()
df_Net_New["num_duration"] = df_Net_New["duration"].str.extract("(\d+)").astype(float)
 # since the duration column has object data type too,we are extracting only the numerical part of the column by creating a new column called "num_duration"
```

**The Measures of Central Tendency are:**  *Mean, Median, Mode*

**The Measures of Dispersion are:** *Variance, Standard Deviation*

```python
Mean = df_Net_New["num_duration"].mean()
Median = df_Net_New["num_duration"].median()
Mode = df_Net_New["num_duration"].mode()
Variance = df_Net_New["num_duration"].var()
Standard_Deviation = df_Net_New["num_duration"].std()
print("The Mean duration is: ",Mean,"min","\nThe Median duration is: ",Median,"min","\nThe Mode of duration is: ",Mode)
print("\nThe Variance of duration is: ",Variance,"min","\nThe Standard Deviation of duration is: ",Standard_Deviation,"min")
```

**Observations:** *On an average the length of movie is around 70 minutes and the standard deviation is 50 which is 20 greater than mean of duration.*
